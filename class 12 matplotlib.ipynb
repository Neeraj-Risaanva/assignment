```python
x = np.array(["A","B","C","D","E"])
y = np.array([5,9,11,13,15])

plt.bar(x,y,color="blue")
plt.show()
```

```python
### If you want the bars to be displayed horizontally instead of vertically, use the barh() functio
x = np.array(["A","B","C","D","E"])
y = np.array([5,9,11,13,15])

plt.barh(x,y,color="green")
plt.show()
```

```python
x = np.array(["A","B","C","D","E"])
y = np.array([5,9,11,13,15])

plt.bar(x,y,color="blue" , width=0.3)
plt.show()
```

```python
### The barh() takes the keyword height to set the height of the bars
x = np.array(["A","B","C","D","E"])
y = np.array([5,9,11,13,15])

plt.barh(x,y,color="blue", height=0.2)
plt.show()
```

```python
y = np.array([50,75,89,12,35])
plt.pie(y)
plt.show()
```

```python
### Add labels to the pie chart with the label parameter.

### The labels parameter must be an array with one label for each wedge
y = np.array([10,16,19,30,35])
mylabels = ["Education","Health","Transport","Tax","Labour"]
plt.pie(y, labels = mylabels)
plt.show()
```

```python
### As mentioned the default start angle is at the x-axis, but you can change the start angle
### by specifying a startangle parameter.

### The startangle parameter is defined with an angle in degrees, default angle is 0:
y = np.array([45,30,25,20])
mylabels = ["Apple","Bananas","Orange","Mango"]

plt.pie(y, labels = mylabels, startangle = 90)
plt.show()
```

```python
y = np.array([45,30,25,20])
mylabels = ["Apple","Bananas","Orange","Mango"]
myexplode =[0.35,0.59,.90,1]

plt.pie(y, labels = mylabels, startangle = 90, explode = myexplode)
plt.show()
```

```python
y = np.array([45,30,25,20])
mylabels = ["Apple","Bananas","Orange","Mango"]
mycolors = ["black","orange","blue","grey"]

plt.pie(y, labels = mylabels, startangle = 90, colors = mycolors)
plt.show()
```
